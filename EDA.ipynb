# EDA

```python
import pandas as pd
import json
import seaborn as sns
from matplotlib import pyplot as plt
```

## Initial EDA for the dev dataset

```python
reviews = []
for line in open('data/dataset_en_dev.json', 'r'):
    reviews.append(json.loads(line))
reviews[0]
```

```python
# Creating a dataframe of this data
df = pd.DataFrame(reviews)
df.info()
```

```python
df.head()
```

```python
# Basic statistics about the data
df.describe()
```

```python
fig, axs = plt.subplots(2, 1, figsize = (20,12))

sns.histplot(df, x = 'stars', ax = axs[0])
sns.histplot(df, x = 'product_category', ax = axs[1])
plt.setp(axs[1].xaxis.get_majorticklabels(), rotation=45, size=9)
plt.show()
```
